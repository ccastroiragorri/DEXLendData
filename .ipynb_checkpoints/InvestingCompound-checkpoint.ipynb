## Returns for Passive Investors - Suppliers

```python
def passiveInv_ETH(USDInv,mofInv, horizon):
    #initial date index
    ip=BR_E.index.get_loc(mofInv, method='nearest')
    ia_ETH=USDInv/PUSD_E.iloc[ip]
    ia_cETH=ia_ETH/Xrate_E.iloc[ip]
    if (ip+horizon > len(BR_E.index)):
        print("Investment Horizon oversteps historical data")
    else:
        o_ETH=ia_cETH*Xrate_E.iloc[ip+horizon]
        ti_inc=(o_ETH-ia_ETH)*PUSD_E.iloc[ip+horizon]
        p_inc=PUSD_E.iloc[ip+horizon]-PUSD_E.iloc[ip]
        ti_ret=(ti_inc/(ia_ETH*PUSD_E.iloc[ip+horizon]))*100
        dp_ret= (p_inc/PUSD_E.iloc[ip])*100
    idt=(USDInv,ia_ETH,ia_cETH,math.nan,math.nan,math.nan,math.nan)
    hdt=((o_ETH*PUSD_E.iloc[ip+horizon]),o_ETH,ia_cETH,ti_inc,p_inc,ti_ret,dp_ret)
    vnames=('USD','ETH','cETH','int. Income','Price Income','int. %','Price %')
    tbl= pd.DataFrame(np.concatenate((idt,hdt)).reshape(2,7).transpose(), index=vnames,columns=["initial",'days'+str(horizon)])
    return tbl.round(4)
```

```python
passiveInv_ETH(1000,date(2019, 12, 15), 360)
```

```python
def passiveInv_DAI(USDInv,mofInv, horizon):
    #initial date index
    ip=BR_D.index.get_loc(mofInv, method='nearest')
    ia_ETH=USDInv/PUSD_D[ip]
    ia_cETH=ia_ETH/Xrate_D[ip]
    if (ip+horizon > len(BR_D.index)):
        print("Investment Horizon oversteps historical data")
    else:
        o_ETH=ia_cETH*Xrate_D[ip+horizon]
        ti_inc=(o_ETH-ia_ETH)*PUSD_D[ip+horizon]
        p_inc=PUSD_D[ip+horizon]-PUSD_D[ip]
        ti_ret=(ti_inc/(ia_ETH*PUSD_D[ip+horizon]))*100
        dp_ret= (p_inc/PUSD_E[ip])*100
    idt=(USDInv,ia_ETH,ia_cETH,math.nan,math.nan,math.nan,math.nan)
    hdt=((o_ETH*PUSD_D[ip+horizon]),o_ETH,ia_cETH,ti_inc,p_inc,ti_ret,dp_ret)
    vnames=('USD','DAI','cDAI','int. Income','Price Income','int. %','Price %')
    tbl= pd.DataFrame(np.concatenate((idt,hdt)).reshape(2,7).transpose(), index=vnames,columns=["initial",'days'+str(horizon)])
    return tbl.round(4)
```

## Borrower ETH

```python
USDm=1000
dinv=date(2019, 12, 15)
horizon=90
safety=0.3
```

```python
def borrower_ETH(USDm,dinv,horizon,safety):
    cf_E=0.75
    ip_E=BR_E.index.get_loc(dinv, method='nearest')
    #initial amount of Collateral 
    ia_ETH=USDm/PUSD_E.iloc[ip_E] #in ether
    ia_cETH=ia_ETH/Xrate_E[ip_E] #in cETH
    #Max Borrow in collateral currency: ETH
    mx_b_ETH=ia_ETH*cf_E
    #Borrow 
    b_ETH=mx_b_ETH*(1-safety) #in ETH keeps a buffer defined by safetty 
    b_USD=b_ETH*PUSD_E.iloc[ip_E] # in USD uses the money outside of protocol
    b_cETH=b_ETH/Xrate_E[ip_E] # in cETH
    Mcalls=sum(mx_b_ETH/(b_USD/PUSD_E[ip_E:ip_E+horizon])<1) #Number of Margin Calls over investment horizon
    pay_ETH=b_cETH*Xrate_E.iloc[ip_E+horizon] #accured interest in form of cETH in ETH for borrowing in ETH
    pay_USD=pay_ETH*PUSD_E.iloc[ip_E+horizon] #borrowed to pay in USD
    loan_cost=(pay_USD-b_USD) #interest payed in USD cost of loan
    ti_cost=(pay_ETH-b_ETH)*PUSD_E.iloc[ip_E+horizon]  #in USD
    p_inc=PUSD_E.iloc[ip_E+horizon]-PUSD_E.iloc[ip_E]
    # Percentages
    ti_ret_cost=(ti_cost/b_USD)*100
    dp_ret= (p_inc/PUSD_E.iloc[ip_E])*100
    idt=(USDm,   ia_ETH,ia_cETH, b_USD, b_ETH, math.nan, math.nan, math.nan, math.nan, math.nan )
    hdt=(math.nan,math.nan, math.nan,  pay_USD, pay_ETH, Mcalls, ti_cost, p_inc, ti_ret_cost, dp_ret)
    vnames=('USD','ETH','cETH','Borrow USD','Borrow ETH', 'MCalls','Int. Pay',  'Price Income','int. pay %','Price %')
    tbl= pd.DataFrame(np.concatenate((idt,hdt)).reshape(2,10).transpose(), index=vnames,columns=["initial",'days'+str(horizon)])
    return tbl.round(4)
```

```python
borrower_ETH(1000,date(2019, 12, 15),90,0.3)
```

```python
borrower_ETH(1000,date(2020, 3, 15),60,0.3)
```

## Borrowing in Compound (margin trader) post collateral in liquid crypto ETH borrow in DAI

```python
def activeInv(USDm,dinv,horizon,safety):
    #collateral market parameter
    cf_E=0.75
    #investment date index
    ip_E=BR_E.index.get_loc(dinv, method='nearest')
    ip_D=BR_D.index.get_loc(dinv, method='nearest')
    #initial amount of Collateral 
    ia_ETH=USDm/PUSD_E.iloc[ip_E] #in ether
    ia_cETH=ia_ETH/Xrate_E[ip_E] #in cETH
    #Max Borrow in collateral currency: ETH
    mx_b_ETH=ia_ETH*cf_E
    #Borrow 
    b_ETH=mx_b_ETH*(1-safety) #in ETH keeps a buffer defined by safetty 
    #r_cETH=ia_cETH-(b_ETH/Xrate_E[ip_E]) # remaning cETH
    CR=((PUSD_E[ip_E]*ia_ETH)/(PUSD_E[ip_E]*b_ETH))*100 #collatreral ratio un USD min 133 
    b_USD=b_ETH*PUSD_E.iloc[ip_E] # in USD
    b_DAI=b_USD/PUSD_D.iloc[ip_D] # in DAI
    b_cDAI=b_DAI/Xrate_D.iloc[ip_D] #in cDAI
    #Use borrow to leverage ETH Portfolio
    add_ETH=b_USD/PUSD_E.iloc[ip_E] #in ETH
    l_ETH=ia_ETH+add_ETH #leveraged portfolio in ETH
    l_cETH=l_ETH/Xrate_E.iloc[ip_E] #leveraged portfolio in cETH
    x_Leverage=l_ETH/ia_ETH # x times leverage
    Mcalls=sum(mx_b_ETH/(b_USD/PUSD_E[ip_E:ip_E+horizon])<1) #Number of Margin Calls over investment horizon
    #Value at the end of investment horizon
    #o_ETH=l_cETH*Xrate_E.iloc[ip_E+horizon] #accured interest from cETH to ETH
    ti_inc=(ia_cETH*Xrate_E.iloc[ip_E+horizon]-ia_ETH)*PUSD_E.iloc[ip_E+horizon] #interest paid in USD for suppling ETH 
    pay_DAI=b_cDAI*Xrate_D.iloc[ip_D+horizon] #accured interest in form of cDAI in DAI for borrowing in DAI
    pay_USD=pay_DAI*PUSD_D.iloc[ip_D+horizon] #borrowed to pay in USD
    ti_cost=(pay_USD-b_USD) #interest payed in USD cost of loan
    benefit=l_ETH*PUSD_E.iloc[ip_E+horizon]-pay_USD #value of investment at end of horizon
    p_inc=PUSD_E.iloc[ip_E+horizon]-PUSD_E.iloc[ip_E]
    # Percentages
    ti_ret_inc=(ti_inc/(ia_ETH*PUSD_E.iloc[ip_E+horizon]))*100
    ti_ret_cost=(ti_cost/b_USD)*100
    dp_ret= (p_inc/PUSD_E.iloc[ip_E])*100
    idt=(USDm,   ia_ETH,ia_cETH, b_USD, x_Leverage, Mcalls,           math.nan, math.nan, math.nan, math.nan )
    hdt=(benefit,l_ETH, l_cETH,  pay_USD, ti_inc,        ti_cost,      p_inc,         ti_ret_inc,  ti_ret_cost, dp_ret)
    vnames=('USD','ETH','cETH','Borrow USD', 'xlev/Int. Income', 'MCalls/Int. Pay',  'Price Income','int. inc %','int. pay %','Price %')
    tbl= pd.DataFrame(np.concatenate((idt,hdt)).reshape(2,10).transpose(), index=vnames,columns=["initial",'days'+str(horizon)])
    return tbl.round(4)
```
